```python
import glob
import pandas as pd
import os

#intialize empty list that we will append dataframes to
os.chdir("E:/2019Fall/6211/A08/mydata/names")
list_data = []

#write a for loop that will go through each of the file name through globbing and the end result will be the list of dataframes
for filename in glob.glob('*.txt'):
    data = pd.read_csv(filename,names='NGC')
    list_data.append(data)

#print list_data which is a list of dataframes
list_data
 
pd.concat(list_data,ignore_index=True,sort=True)
```

```python
#Q1
#Open the file and append all records to the list
mylist = []

with open("E:/2019Fall/6211/A08/mydata/names/yob1880.txt", "r") as f1:
    content1 = f1.readlines()
for line in content1:
    oneline = line.strip()
    linelist = oneline.split(',')
    namelist = []
    for n in linelist:
        namelist.append(n)
    mylist.append(namelist)
    
with open("E:/2019Fall/6211/A08/mydata/names/yob1980.txt", "r") as f2:
    content2 = f2.readlines()
for line in content2:
    oneline = line.strip()
    linelist = oneline.split(',')
    namelist = []
    for n in linelist:
        namelist.append(n)
    mylist.append(namelist)
    
mylist_new = [record for record in mylist if int(record[2]) > 100 and int(record[2]) <= 5000]
mylist_new
```

```python
#Q2. 
name_A = []   # create a list containing baby names that start with the letter "A"
for record in mylist:
    if record[0][0] == 'A':    #start with the letter "A"
        name_A.append(record)

#Calculate the average number of births for names that start with the letter “A”
sum_A = 0
for n in name_A:
    sum_A = sum_A + int(n[2])

avg_A = sum_A / len(name_A)
avg_A
```

```python
#Q3. 
name_a = []   # create a list containing baby names that end with the letter "a"
for record in mylist:
    if record[0][-1] == 'a':       #end with the letter "a"
        name_a.append(record)

#Calculate the average number of births for names that end with the letter “a”
sum_a = 0
for n in name_a:
    sum_a = sum_a + int(n[2])

avg_a = sum_a / len(name_a)
avg_a
```
